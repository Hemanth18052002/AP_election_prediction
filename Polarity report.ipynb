```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
camp_ysrcp = pd.read_csv(r"dataset\campagin_speeches\camp_ysrcp_30_10_2023_23_54_15.csv")
camp_tdp = pd.read_csv(r"dataset\campagin_speeches\camp_tdp_30102023_23_54_25.csv")

mani_ysrcp = pd.read_csv(r"dataset\manifestios\mani_ysrcp_31_10_2023_00_07_04.csv")
mani_tdp = pd.read_csv(r"dataset\manifestios\mani_tdp_31102023_00_07_07.csv")
```

```python
mani_ysrcp_total = sum(mani_dic_ysrcp.values())
mani_tdp_total = sum(mani_dic_tdp.values())
mani_x2 = ['Party Y Manifestos','Party T Manifestos']
mani_x1 = {'Positive':[round((mani_dic_ysrcp['Positive']/mani_ysrcp_total)*100,2)
                       , round((mani_dic_tdp['Positive']/mani_tdp_total)*100,2)],
          'Negative':[round((mani_dic_ysrcp['Negative']/mani_ysrcp_total)*100,2), 
                      round((mani_dic_tdp['Negative']/mani_tdp_total)*100,2)],
          'Neutral':[round((mani_dic_ysrcp['Netural']/mani_ysrcp_total)*100,2), 
                     round((mani_dic_tdp['Neutral']/mani_tdp_total)*100,2)]}
mani_df = pd.DataFrame(mani_x1,mani_x2)
mani_df
```

```python
ysrcp_camp = ((camp_dic_ysrcp['Positive']/camp_ysrcp_total)+
             (camp_dic_tdp['Negative']/camp_tdp_total)+
             (camp_dic_ysrcp['Neutral']/camp_ysrcp_total)+
             (camp_dic_tdp['Neutral']/camp_tdp_total))/((camp_dic_ysrcp['Positive']/camp_ysrcp_total)+
                                                       (camp_dic_tdp['Positive']/camp_tdp_total)+
                                                       (camp_dic_ysrcp['Negative']/camp_ysrcp_total)+
                                                       (camp_dic_tdp['Negative']/camp_tdp_total)+
                                                       (camp_dic_ysrcp['Neutral']/camp_ysrcp_total)+
                                                       (camp_dic_tdp['Neutral']/camp_tdp_total))

tdp_camp = ((camp_dic_tdp['Positive']/camp_tdp_total)+
           (camp_dic_ysrcp['Negative']/camp_ysrcp_total)+
           (camp_dic_ysrcp['Neutral']/camp_ysrcp_total)+
             (camp_dic_tdp['Neutral']/camp_tdp_total))/((camp_dic_ysrcp['Positive']/camp_ysrcp_total)+
                                                       (camp_dic_tdp['Positive']/camp_tdp_total)+
                                                       (camp_dic_ysrcp['Negative']/camp_ysrcp_total)+
                                                       (camp_dic_tdp['Negative']/camp_tdp_total)+
                                                       (camp_dic_ysrcp['Neutral']/camp_ysrcp_total)+
                                                       (camp_dic_tdp['Neutral']/camp_tdp_total))




ysrcp_mani = ((mani_dic_ysrcp['Positive']/mani_ysrcp_total)+
             (mani_dic_tdp['Negative']/mani_tdp_total)+
             (mani_dic_ysrcp['Netural']/mani_ysrcp_total)+
             (mani_dic_tdp['Neutral']/mani_tdp_total))/((mani_dic_ysrcp['Positive']/mani_ysrcp_total)+
                                                       (mani_dic_tdp['Positive']/mani_tdp_total)+
                                                       (mani_dic_ysrcp['Negative']/mani_ysrcp_total)+
                                                       (mani_dic_tdp['Negative']/mani_tdp_total)+
                                                       (mani_dic_ysrcp['Netural']/mani_ysrcp_total)+
                                                       (mani_dic_tdp['Neutral']/mani_tdp_total))

tdp_mani = ((mani_dic_tdp['Positive']/mani_tdp_total)+
           (mani_dic_ysrcp['Negative']/mani_ysrcp_total)+
           (mani_dic_ysrcp['Netural']/mani_ysrcp_total)+
             (mani_dic_tdp['Neutral']/mani_tdp_total))/((mani_dic_ysrcp['Positive']/mani_ysrcp_total)+
                                                       (mani_dic_tdp['Positive']/mani_tdp_total)+
                                                       (mani_dic_ysrcp['Negative']/mani_ysrcp_total)+
                                                       (mani_dic_tdp['Negative']/mani_tdp_total)+
                                                       (mani_dic_ysrcp['Netural']/mani_ysrcp_total)+
                                                       (mani_dic_tdp['Neutral']/mani_tdp_total))
```

```python
# PUBLIC POLLS
ysrcp_public_polls = [19,15,22,18,15,24,24] 
tdp_public_polls = [6,10,3,7,10,1,1]

ysrcp_public_polls_x = ((100/((sum(ysrcp_public_polls)/len(ysrcp_public_polls))+
                            (sum(tdp_public_polls)/len(tdp_public_polls))))*
                                (sum(ysrcp_public_polls)/len(ysrcp_public_polls)))
tdp_public_polls_y = ((100/((sum(ysrcp_public_polls)/len(ysrcp_public_polls))+
                            (sum(tdp_public_polls)/len(tdp_public_polls))))*
                                 (sum(tdp_public_polls)/len(tdp_public_polls)))

                                  
# CAMPAIGN SPEECHES
ysrcp_camp_x = (100/(ysrcp_camp+tdp_camp))*(ysrcp_camp)
tdp_camp_y = (100/(ysrcp_camp+tdp_camp))*(tdp_camp)

                                  
# MANIFESTOS
ysrcp_mani_x = (100/(ysrcp_mani+tdp_mani))*(ysrcp_mani)
tdp_mani_y = (100/(ysrcp_mani+tdp_mani))*(tdp_mani)                         
```

```python
result_x1 = {'Party Y':[ysrcp_public_polls_x, ysrcp_camp_x, ysrcp_mani_x],
            'Party T':[tdp_public_polls_y, tdp_camp_y, tdp_mani_y]}
result_x2 = ['PUBLIC POLLS', 'CAMPAIGN SPEECHES', 'MANIFESTOS']
result = pd.DataFrame(result_x1,result_x2)
result
```

# WINNING PREDICTION

```python
party = ['Party Y','Party T']
support = []

ysrcp_sup = (ysrcp_public_polls_x + ysrcp_camp_x + ysrcp_mani_x)/3
tdp_sup = (tdp_public_polls_y + tdp_camp_y + tdp_mani_y)/3
support.append((100/(ysrcp_sup+tdp_sup))*ysrcp_sup)
support.append((100/(ysrcp_sup+tdp_sup))*tdp_sup)


if support[0] > support[1]:
    print(f'{party[0]} party have more chance to win in 2024 elections in Andhra Pradesh')
elif support[1] > support[0]:
    print(f'{party[1]} party have more chance to win in 2024 elections in Andhra Pradesh')
else:
    print(f'{party[0]} and {party[1]} party have equal chance to win in 2024 elections in Andhra Pradesh')

print(f'With Majority {abs(support[0]-support[1])}')
```
